```python
#使用python的request模組去訪問https://jsonplaceholder.typicode.com/users
import requests
import datetime
import json
import os
x=0
data=[]
#=================================================
r = requests.get('https://jsonplaceholder.typicode.com/users')
#取得第2,3,4,5 筆資料
#將內容分段存入
print('='*50)
for part in r.json()[1:5]:
    #將資料格式化輸出
    print('\nid\t',part['id'],'\nname\t',part['name'],'\nusername\t',part['username'],'\nemail\t',part['email'])
    print('phone\t',part['phone'],'\nwebsite\t',part['website'],'\ncompany\t',part['company'],'\naddress\t',part['address'],'\n','='*50)
    data.append(json.dumps(r.json()[x+1]))
    x=x+1

#=================================================
#取得當天日期
nowd=datetime.datetime.utcnow().strftime('%Y-%m-%d')

#=================================================
#將當天日期作為檔名開頭
filename=nowd+"-json.txt"

#=================================================
#確定目錄是否存在，若無則新增
path = "./savedata/"
if not os.path.isdir(path):
    os.mkdir(path)

#=================================================
#將2345筆資料內容寫入檔案中
openfile = open('./savedata/'+str(filename),'w')

# 將jsonf1寫入testjson檔案中
for i in range(0,4):
    openfile.write(data[i]+"\n\n")
    
# 將檔案testjson關閉
openfile.close()
#=================================================
print(filename,"存檔完成~~")
```
